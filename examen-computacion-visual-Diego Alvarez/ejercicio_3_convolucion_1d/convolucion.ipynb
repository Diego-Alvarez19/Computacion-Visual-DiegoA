```python
import numpy as np
import matplotlib.pyplot as plt
import os

# 1. Función de convolución manual
def convolucion_1d(senal, kernel):
    n = len(senal)
    k = len(kernel)
    resultado = np.zeros(n - k + 1)

    for i in range(len(resultado)):
        for j in range(k):
            resultado[i] += senal[i + j] * kernel[j]

    return resultado

# 2. Crear datos
np.random.seed(42)
senal = np.random.randint(10, 20, 15)  # 15 valores entre 10-20
kernel = np.array([1, 0, -1])  # Kernel de tamaño 3

# 3. Calcular convolución
resultado = convolucion_1d(senal, kernel)

# 4. Mostrar gráficos en pantalla
plt.figure(figsize=(15, 10))

# Señal original
plt.subplot(3, 1, 1)
markerline, stemlines, _ = plt.stem(senal)
plt.setp(stemlines, 'linewidth', 2)
plt.title(f'Señal Original ({len(senal)} puntos)')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid(alpha=0.3)

# Kernel
plt.subplot(3, 1, 2)
markerline, stemlines, _ = plt.stem(kernel, linefmt='orange')
plt.setp(stemlines, 'linewidth', 2)
plt.title(f'Kernel ({len(kernel)} puntos)')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid(alpha=0.3)

# Resultado
plt.subplot(3, 1, 3)
markerline, stemlines, _ = plt.stem(resultado, linefmt='green')
plt.setp(stemlines, 'linewidth', 2)
plt.title(f'Resultado de Convolución ({len(resultado)} puntos)')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# 5. Guardar gráficos en estructura de archivos
os.makedirs('ejercicio_3_convolucion_1d/graficos', exist_ok=True)

# Señal original
plt.figure(figsize=(10, 4))
plt.stem(senal)
plt.title('Señal Original')
plt.savefig('ejercicio_3_convolucion_1d/graficos/senal_original.png')
plt.close()

# Kernel
plt.figure(figsize=(4, 3))
plt.stem(kernel)
plt.title('Kernel')
plt.savefig('ejercicio_3_convolucion_1d/graficos/kernel.png')
plt.close()

# Resultado
plt.figure(figsize=(10, 4))
plt.stem(resultado)
plt.title('Resultado de la Convolución')
plt.savefig('ejercicio_3_convolucion_1d/graficos/resultado.png')
plt.close()

# 6. Mostrar valores
print("Señal original:", senal)
print("Kernel:", kernel)
print("Resultado:", resultado)
print("Gráficos guardados en 'ejercicio_3_convolucion_1d/graficos/'")
```
